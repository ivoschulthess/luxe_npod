```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# plot settings, increase font sizes
plt.rcParams['figure.figsize'] = 6, 4
plt.rcParams['font.size'] = 12
plt.rcParams['legend.fontsize'] = 11
plt.rcParams['legend.labelspacing'] = 0.2
```

```python
ls_projections = [3,2]

fig, ax = plt.subplots(figsize=(6,5))

# naturalness line
m_a = np.logspace(np.log10(2.3e-2), np.log10(3e0), 1000)
ax.plot(m_a, 1 / (4e5 * (1)**2 * 0.2/m_a), '-.', c='#663300')
ax.text(1, 8e-6, 'naturalness', c='#663300', rotation=18, size=10)

# NA64
data = np.genfromtxt('limit_data/constraints/NA64_2020.txt', delimiter=', ').T
ax.plot(*data, '-', lw=1, c='#333333', label='NA64')
ax.fill(*data, c=ax.get_lines()[-1].get_color(), alpha=0.2)
ax.text(2.4e-2, 3e-4, 'NA64', c='#333333', size=10)

# PrimEx
data = np.genfromtxt('limit_data/constraints/PrimEx_2019.txt').T
ax.plot(*data, '-', lw=1, c='#333333')
ax.fill(*data, c=ax.get_lines()[-1].get_color(), alpha=0.2)
ax.text(1.1e-1, 7e-4, 'PrimEx', c='#333333', rotation=0, size=10)

# OPAL (LEP)
data = np.genfromtxt('limit_data/constraints/OPAL_2017.txt').T
ax.plot(data[0]/1e9, data[1], '-', lw=1, c='#333333')
ax.fill(data[0]/1e9, data[1], c=ax.get_lines()[-1].get_color(), alpha=0.2)
ax.text(6e-2, 8.5e-4, 'OPAL', c='#333333', rotation=0, size=10)

# Belle-II
data = np.genfromtxt('limit_data/constraints/Belle2_2020.txt').T
#ax.plot(data[0], data[1], '-', lw=1, c='#333333')
ax.plot(data[0], data[2], '-', lw=1, c='#333333')
ax.fill(data[0], data[2], c=ax.get_lines()[-1].get_color(), alpha=0.2)
ax.text(6e-1, 3e-4, 'Belle-II', c='#333333', rotation=0, size=10)

# BES III
data = np.genfromtxt('limit_data/constraints/BESIII_2023.txt').T
ax.plot(data[0]/1e9, data[1], '-', lw=1, c='#333333')
ax.fill(data[0]/1e9, data[1], c=ax.get_lines()[-1].get_color(), alpha=0.2)
ax.text(2.2e-1, 2.3e-4, 'BESIII', c='#333333', rotation=0, size=10)

# Beam Dump AxionLimits
data = np.genfromtxt('limit_data/constraints/BeamDump.txt').T
ax.plot(data[0]/1e9, data[1], '-', lw=1, c='#333333')
ax.fill(data[0]/1e9, data[1], c=ax.get_lines()[-1].get_color(), alpha=0.2)
ax.text(2.5e-2, 1.2e-6, 'Beam Dumps', c='#333333', rotation=0, size=10)

data = np.genfromtxt('limit_data/constraints/MiniBooNE.txt').T
ax.plot(data[0]/1e9, data[1], '-', lw=1, c='#333333')
ax.fill(data[0]/1e9, data[1], c=ax.get_lines()[-1].get_color(), alpha=0.2)
ax.text(3e-2, 1e-7, 'MiniBooNE', c='#333333', rotation=-20, size=10)

# FASER
data = np.genfromtxt('limit_data/constraints/FASER.txt').T
ax.plot(data[0]/1e9, data[1], '-', lw=1, c='#333333')
ax.fill(data[0]/1e9, data[1], c=ax.get_lines()[-1].get_color(), alpha=0.2)
ax.text(2.4e-2, 1.1e-4, 'FASER', c='#333333', rotation=-15, size=10)

#####################
# PROJECTIONS
#####################

# Belle-II
data = np.genfromtxt('limit_data/projections/Belle2a.txt').T
ax.plot(*data, 'C2-', lw=1, dashes=ls_projections)
ax.text(6e-1, 3e-5, 'Belle-II', c='C2', size=10)
data = np.genfromtxt('limit_data/projections/Belle2b.txt').T
ax.plot(*data, 'C2-', lw=1, dashes=ls_projections)
#ax.text(7e-2, 1.2e-5, 'Belle-II', c='g', size=10)

# NA62-dump
data = np.genfromtxt('limit_data/projections/NA62_dump_1e18_POT.txt').T
ax.plot(*data, '-', lw=1, c='C3', dashes=ls_projections)
#ax.text(1.7e-1, 5e-7, 'NA62 10¹⁸ POT', c='#009999', size=10)
ax.text(1.7e-1, 5e-7, 'NA62', c='C3', size=10)

# GlueX (check with Yotam)
data = np.genfromtxt('limit_data/projections/reach_gluex_fb.txt').T
ax.plot(*data, 'C5-', lw=1, dashes=ls_projections)
ax.text(5e-1, 1e-4, 'GlueX', c='C5', size=10)

# PrimEx (check with Yotam)
data = np.genfromtxt('limit_data/projections/reach_primex_all.txt').T
ax.plot(*data, '-', lw=1, c='C6', dashes=ls_projections)
ax.text(7.5e-2, 4.2e-4, 'PrimEx', c='C6', size=10)

# FASER (check with Yotam)
data = np.genfromtxt('limit_data/projections/FASER.txt').T
ax.plot(*data, '-', lw=1, c='C1', dashes=ls_projections)
ax.text(2.4e-2, 6.6e-5, 'FASER', c='C1', rotation=-15, size=10)

# FASER2 (check with Yotam)
data = np.genfromtxt('limit_data/projections/FASER2.txt').T
ax.plot(*data, '-', lw=1, c='C1', dashes=ls_projections)
ax.text(3.3e-1, 9e-6, 'FASER2', c='C1', size=10)

# NA64 5e12 EOT
data = np.genfromtxt('limit_data/projections/NA64_5e12_EOT.txt').T
ax.plot(*data, 'C4', lw=1, dashes=ls_projections)
#ax.text(2.4e-2, 3.7e-5, r'NA64 5$\times$10¹² EOT', c='m', size=10)
ax.text(2.4e-2, 3.6e-5, r'NA64', c='C4', size=10)

# SHiP
data = np.genfromtxt('limit_data/projections/SHiP.txt', delimiter=', ').T
ax.plot(*data, c='C0', lw=1, dashes=ls_projections)
ax.text(9e-1, 5e-8, r'SHiP', c='C0', size=10)

#####################
# LUXE
#####################

# LUXE-NPOD phase-0
data = np.genfromtxt('limit_data/projections/luxe_phase0.txt').T
ax.plot(*data, 'b-', lw=2)
ax.text(3e-2, 1e-5, 'LUXE phase-0', c='b', rotation=-15, size=10)

# LUXE NPOD phase-1
data = np.genfromtxt('limit_data/projections/luxe_phase1.txt').T
ax.plot(*data, 'k-', lw=2)
ax.text(3e-2, 3e-6, 'LUXE phase-1', c='k', rotation=-15, size=10)

ax.set(xscale='log', xlabel='$m_a$ [GeV]',  
       yscale='log', ylabel='$g_a$ [GeV$^{-1}$]', 
       xlim=(m_a.min(), 3), ylim=(2e-8, 3.1e-3),
       xticks=(0.05, 0.1, 0.2, 0.5, 1, 2), xticklabels=('0.05','0.1','0.2','0.5','1','2')
      )

#ax.grid(which='both')

fig.set(dpi=200)
plt.show()
```
